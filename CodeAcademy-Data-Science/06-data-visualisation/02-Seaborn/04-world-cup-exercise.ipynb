```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

# load WorldCupMatches.csv
df = pd.read_csv('WorldCupMatches.csv')

# Create a new column'Total Goals', and set it equal to the sum of the columns Home Team Goals and Away Team Goals.
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']

# Set the style of your plot to be whitegrid
sns.set_style("whitegrid")

# set the context to be 'poster'
sns.set_context('poster', font_scale = .4)

# create figure and axes for the plot
f, ax = plt.subplots(figsize=(12,7))

# set title
ax.set_title('World Cup Goals 1930 -2014')

# create a bar chart visualizing how many goals were scored each year the World Cup was held between 1930-2014.
sns.barplot(
	data=df,
  x='Year',
  y='Total Goals'
)

plt.show()
```

```python
# load goals.csv
df_goals = pd.read_csv('goals.csv')
print(df_goals.head())

f, ax2 = plt.subplots(figsize=(12,7))
ax2 = sns.boxplot(
		data=df_goals,
  	x='year',
  	y='goals',
  	palette='Spectral'
)
ax2.set_title('Avg Goals per Year')
plt.show()
```
