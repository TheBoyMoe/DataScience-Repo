# Barchart with Error Bars

```python
from matplotlib import pyplot as plt
import numpy as np

past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]
```

```python
# create a figure of width 10 and height 8.
plt.figure(figsize=(10,8))
```

```python
# Plot the blue bars, which have the heights listed in past_years_averages.
plt.bar(range(len(years)), past_years_averages)
```

```python
# Add error bars of cap size 5 and heights corresponding to the list error.
plt.bar(range(len(years)), past_years_averages, yerr=error, capsize=5)

# Set the axis to go from -0.5 to 6.5 on the x-axis and 70 to 95 on the y-axis.
plt.axis([-0.5,6.5,70,95])
```

```python
plt.bar(range(len(years)), past_years_averages, yerr=error, capsize=5)
plt.axis([-0.5,6.5,70,95])

# Create an ax object using plt.subplot(). Use ax to set the x-axis ticks to be 
# range(len(years)) and the x-axis labels to be the years list.
ax = plt.subplot()
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)

# add title label the axes
plt.title('Final Exam Averages')
plt.xlabel('Year')
plt.ylabel('Test average')

# save the figure to a file
plt.savefig('my_bar_chart.png')

plt.show()
```

# Side-by-side Bars

```python
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]
# school_a_x = [0.8, 2.8, 4.8, 6.8, 8.8]
# school_b_x = [1.6, 3.6, 5.6, 7.6, 9.6]
# Make your chart here
# use create_x to determine where to put bars for 2 data sets
school_a_x = create_x(2, 0.8, 1, len(middle_school_a))
school_b_x = create_x(2, 0.8, 2, len(middle_school_b))

# create figure
plt.figure(figsize=(10,8))

# create a set of axes
ax = plt.subplot()

# plot schools
plt.bar(school_a_x, middle_school_a)
plt.bar(school_b_x, middle_school_b)

# Create a new list of x-values called middle_x, which are the values in the middle of school_a_x and school_b_x. This is where we will place the x-ticks. 
middle_x = [ (a + b) / 2.0 for a, b in zip(school_a_x, school_b_x)]

# set xticks and x-tick labels
ax.set_xticks(middle_x)
ax.set_xticklabels(unit_topics)

# create legend, title and labels
plt.legend(['Middle School A', 'Middle School B'])
plt.title('Test Averages on Different Units')
plt.ylabel('Test Average')
plt.xlabel('Unit')
plt.axis([0,10,70,90])
plt.savefig('my_side_by_side.png')
plt.show()
```

# Stacked Bars

```python
unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

x = range(5)

c_bottom = np.add(As, Bs)
#create d_bottom and f_bottom here
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)
#create your plot here
plt.figure(figsize=(10,8))

# plot As, Bs, etc
col_number = range(len(unit_topics))
plt.bar(col_number, As)
plt.bar(col_number, Bs, bottom=As)
plt.bar(col_number, Cs, bottom=c_bottom)
plt.bar(col_number, Ds, bottom=d_bottom)
plt.bar(col_number, Fs, bottom=f_bottom)

# set the xticks and x-tick labels
ax = plt.subplot()
ax.set_xticks(col_number)
ax.set_xticklabels(unit_topics)

# set title and axes labels
plt.title('Grade distribution')
plt.xlabel('Unit')
plt.ylabel('Number of Students')

plt.savefig('my_stacked_bar.png')
plt.show()
```
