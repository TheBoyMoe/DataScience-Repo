### Frequency counts for categorical data

As you've seen, `.describe()` can only be used on numeric columns. So how can you diagnose data issues when you have categorical data? One way is by using the `.value_counts()` method, which returns the frequency counts for each unique value in a column!

This method also has an optional parameter called `dropna` which is `True` by **default**. What this means is if you have missing data in a column, it will not give a frequency count of them. You want to **set the dropna column to False** so if there are missing values in a column, it will give you the frequency counts.

Check the `Borough`, `State`, and `Site Fill` columns to make sure all the values in there are valid. When looking at the output, do a sanity check: Are all values in the `State` column from NY, for example? Since the dataset consists of applications filed in NY, you would expect this to be the case.

```python
df_subset.Borough.value_counts(dropna=False)
```

```python
df.State.value_counts(dropna=False)
```

```python
df['Site Fill'].value_counts(dropna=False)
```

### Visualizing Data

Is a great way to spot outliers and obvious errors. 

Use `histograms` and `barplots` to look at the frequencies of data, which can be used to look for potential errors and outliers. Not all outliers are bad values, some may be valid.

#### Barplots

Used for discreate data counts (for any value in a range, there is a positive minimum distance to the next value in the range, e.g. variables that can only be integers, or a variable that can be only 0 or 1)

- visualize basic summary statistics into a single figure that can be used to compare across multiple categories.
- create the plot by calling the `boxplot()` method on the dataframe, e.g. `df.boxplot(column='population', by='continent)` - compare the population of each column.
- the ends of the 'whiskers' mark the min/max values, excluding any outliers. The outliers are shown beyond the whiskers.

#### Histograms

Used for continuous data counts (a value that can take up any value within a range). 

- plot the histogram on the column of interest, e.g. `df.population.plot('hist')`
- y-axis shows how many values in the data are in a particular range of values(the frequency)
- x-axis shows the range of values that are counted

#### Scatter Plots

Used to plot the relationship between two numeric variables/columns.

- you can use this to flag potentially bad data.
- can show errors not found by looking at histograms or boxplots(plots of a single variable)

### Visualizing single variables with histograms

The `.plot()` method allows you to create a plot of each column of a DataFrame. The `kind` parameter allows you to specify the type of plot to use, `kind='hist'`, for example, plots a histogram.

Begin by computing summary statistics for the `Existing Zoning Sqft` column using the `.describe()` method. You'll notice that there are extremely large differences between the `min` and `max` values, and the plot will need to be adjusted accordingly. In such cases, it's good to look at the plot on a `log` scale. The keyword arguments `logx=True` or `logy=True` can be passed in to `.plot()` depending on which axis you want to rescale.

Finally, note that Python will render a plot such that the axis will hold all the information. That is, if you end up with large amounts of whitespace in your plot, it indicates counts or values too small to render.

```python
import matplotlib.pyplot as plt

df_subset['Existing Zoning Sqft'].describe()
```

Create a histogram of the `Existing Zoning Sqft` column. Rotate the axis labels by 70 degrees and use a log scale for both axes.

```python
df_subset['Existing Zoning Sqft'].plot(kind='hist', rot=70, logx=True, logy=True)

plt.show()
```

You expected a large amount of counts on the left side of the plot because the 25th, 50th, and 75th percentiles have a value of `0`. The plot shows us that there are barely any counts near the `max` value, signifying an outlier.

### Visualizing multiple variables with boxplots

Histograms are great ways of visualizing **single variables**. To visualize **multiple variables**, boxplots are useful, especially when one of the variables is categorical.

We'll want to create a new column, `inital_cost`, with the inital cost as a numerical value with the '$' symbol removed.

```python
# use regex to turn 'Inital Cost' column into numerical type
df_subset['initial_cost'] = df_subset['Initial Cost'].replace('[\$,]', '', regex=True)
df_subset['initial_cost'] = pd.to_numeric(df_subset['initial_cost'])
df_subset.head()
```

Use a `boxplot` to compare the `initial_cost` across the different values of the `Borough` column. The pandas `.boxplot()` method is a quick way to do this, in which you have to specify the column and by parameters. Here, you want to visualize how `initial_cost` varies by `Borough`.

```python
df_subset.boxplot(column='initial_cost', by='Borough', rot=90)
plt.show()
```

You can see the 2 extreme outliers are in the borough of Manhattan. An initial guess could be that since land in Manhattan is extremely expensive, these outliers may be valid data points. Again, further investigation is needed to determine whether or not you can drop or keep those points in your data.

### Visualizing multiple variables with scatter plots

Boxplots are great when you have a numeric column that you want to compare across different categories. When you want to visualize two numeric columns, scatter plots are ideal.

We'll want to create a new column, `total_est_fee`, with the `Total Est Fee` as a numerical value with the '$' symbol removed.

```python
# use regex to turn 'Total Est Fee' column into numerical type
df_subset['total_est_fee'] = df_subset['Total Est. Fee'].replace('[\$,]', '', regex=True)
df_subset['total_est_fee'] = pd.to_numeric(df_subset['total_est_fee'])
df_subset.head()
```

Make a scatter plot with `initial_cost` on the x-axis and the `total_est_fee` on the y-axis. You can do this by using the DataFrame `.plot()` method with `kind='scatter'`. You'll notice right away that there are 2 major outliers shown in the plots.

Since these outliers dominate the plot, an additional DataFrame, df_subset, has been provided, in which some of the extreme values have been removed. After making a scatter plot using this, you'll find some interesting patterns here that would not have been seen by looking at summary statistics or 1 variable plots. See image below of updated plot

```python
# Create and display the first scatter plot
df_subset.plot(kind='scatter', x='initial_cost', y='total_est_fee', rot=70)
plt.show()
```

In general, from the second plot it seems like there is a strong correlation between `initial_cost` and `total_est_fee`. In addition, take note of the large number of points that have an `initial_cost` of `0`. It is difficult to infer any trends from the first plot because it is dominated by the outliers.
