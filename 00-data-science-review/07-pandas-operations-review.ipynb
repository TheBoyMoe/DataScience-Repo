```python
# df.groupby('Survived').PassengerId.count() # 1 == survived
df.Survived.value_counts()
```

```python
# df.groupby('Pclass').PassengerId.count()
df.Pclass.value_counts()
```

```python
# numbers that survived
df.groupby(['Survived', 'Pclass']).PassengerId.count()
```

```python
# percentages that survived
df[['Pclass', 'Survived']].groupby('Pclass', as_index=False).mean()
```

```python
# df.groupby('Embarked').PassengerId.count()
df.Embarked.value_counts()
```

```python
# numbers that Survived
df.groupby(['Survived', 'Embarked']).PassengerId.count()
```

```python
# percentage that survived
df[['Embarked', 'Survived']].groupby('Embarked', as_index=False).mean()
```
