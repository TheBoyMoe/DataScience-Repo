### Visual EDA

```python
# plot a scatter-matrix
# c - color(used target variable so each pop is colored differently), s - marker size
pd.scatter_matrix(df, c=y, figsize=(8,9), s=150, marker='D');
```

There is a strong correlation between petal length and width.

On this particular example a `scatter-matrix` makes sense when visualizing the dataset. However, when our features are either `0` or `1` a different type of plot would be more useful here, such as Seaborn's `countplot`.

```python
data = pd.read_csv('../datasets/house-votes-84.csv', header=None)
data.columns = ['party', 'infants', 'water', 'budget', 'physician', 'salvador',
       'religious', 'satellite', 'aid', 'missile', 'immigration', 'synfuels',
       'education', 'superfund', 'crime', 'duty_free_exports', 'eaa_rsa']
data.head().T
```

```python
data.info()
```

```python
# visuaizing the 'education' bill votes
plt.figure()
sns.countplot(x='education', hue='party', data=data, palette='RdBu')
plt.xticks([0,1], ['No', 'Yes'])
plt.show()
```

In `sns.countplot()`, we specify the x-axis data to be `education`, and hue to be `party`. Recall that `party` is also our target variable. So the resulting plot shows the difference in voting behavior between the two parties for the `education` bill, with each party colored differently. We manually specified the color to be `RdBu`.

Be sure to begin your plotting statements for each figure with `plt.figure()` so that a new figure will be set up. Otherwise, your plots will be overlayed onto the same figure.

```python
# examining the 'satellite' bill vote
plt.figure()
sns.countplot(x='satellite', hue='party', data=data, palette='RdBu')
plt.xticks([0,1], ['No', 'Yes'])
plt.show()
```

```python
# examining the 'missile' bill votes
plt.figure()
sns.countplot(x='missile', hue='party', data=data, palette='RdBu')
plt.xticks([0,1], ['No', 'Yes'])
plt.show()
```
